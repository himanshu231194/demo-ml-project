```python
# Step 1: Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pickle
```

```python
# Step 2: Prepare the dataset

# Sample dataset
data = {
    'cgpa': [6.5, 7.0, 7.2, 7.5, 8.0, 8.5, 9.0],
    'placement_package_lpa': [4.0, 4.2, 4.3, 5.0, 5.4, 5.9, 6.3]
}

df = pd.DataFrame(data)

# Define features and labels
x = df[['cgpa']]  # 2D
y = df['placement_package_lpa']  # 1D
```

```python
# Step 3: Split into training and testing

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=19
)
```

```python
# Step 4: Train the model

lr = LinearRegression()
lr.fit(x_train, y_train)
pickle.dump(lr, open("simple_linear_model.pkl", "wb"))
```

```python
# Step 5: Predict and score

y_pred = lr.predict(x_test)

# Check how good the model is
score = lr.score(x_test, y_test)
print("R² score:", round(score, 2))
print("Slope (m):", lr.coef_[0])
print("Intercept (c):", lr.intercept_)
score
```

```python
# Step 6: Plot the line

plt.scatter(x, y, color='blue', label='Actual data')
plt.plot(x, lr.predict(x), color='red', label='Regression Line')
plt.xlabel('CGPA')
plt.ylabel('Package (LPA)')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```
