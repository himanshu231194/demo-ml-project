```python
# y = mx + c
# y = dependent variable
# x = independent variable
# m = slope/gradient/coefficent
# c = intercept => if negation then mean line cut y axis below 0 and vicevars


# create dataset
# check head(5)
# check isnull().sum()
# now check dependent and independent variable and filter out
# check summary() for mean, median, mode, std, var, count
# get x (2D) and y (1D)
# create graph to check if it follow linear relationship
# training and testing and train_test_spliting 
# lr.fit() => find out the best value of m and c (fit is use to train the modal based on m and c )
# It finds the best values for m and c such that the line y = mx + c fits the training data as closely as possible.    This is done by minimizing the error (difference) between actual y_train and predicted y.

# create dataset
# create dataset


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

data = {
    'cgpa': [6.5, 7.2, 7.5, 7.8, 8.0, 8.2, 8.4],
    'package': [4.0, 4.2, 5.0, 5.5, 5.2, 6.0, 6.5]
}


dataset = pd.DataFrame(data)
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
# should be 2D
x = dataset[['cgpa']] 
y = dataset['package']
```

```python
sns.scatterplot(x='cgpa', y='package', data=dataset)
plt.show()
# as graph is moving in same direction it means it follow linearity
```

```python
# training and testing
# x	Independent variable(s) — features (e.g., CGPA, experience)
# y	Dependent variable — target (e.g., salary, placement package)
# train_test_split	A function from sklearn.model_selection that randomly splits the data
# test_size=0.2	20% of the data goes to the test set, 80% to the training set
# random_state=42	Ensures the split is reproducible (you’ll get the same split every time)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=15)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
lr.predict([[6.5]])
```

```python
plt.scatter(x, y, color='blue', label='Actual data')
plt.plot(x, lr.predict(x), color='red', label='Regression Line')
plt.xlabel('CGPA')
plt.ylabel('Package (LPA)')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```
